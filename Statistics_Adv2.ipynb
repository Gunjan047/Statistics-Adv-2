**Question 1: What is hypothesis testing in statistics?**

**Answer** - Hypothesis testing is a statistical method used to evaluate a claim or assumption about a population by analyzing data from a sample. This formal procedure allows researchers to determine if their findings are statistically significant or likely to have occurred by chance.

**Real-Life Example**

A company claims its new battery lasts 10 hours. You test a sample and find the average is 9.5 hours. Hypothesis testing helps you decide whether this difference is statistically significant or just due to random variation.

**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer**- The significance level, or alpha (Î±), is a crucial threshold set before a hypothesis test begins. It defines the maximum probability of a Type I error, which is incorrectly rejecting a true null hypothesis.

Â What the Significance Level Represents

 - **Type I error risk:** The significance level represents the acceptable risk of a "false positive" conclusion. For instance, if (Î±) = 0.05, there is a 5% chance of rejecting the null hypothesis when it is true.

 - **Decision threshold:** The significance level creates a "rejection region" on the distribution curve of a test statistic. Any test result that falls into this region is considered unusual enough to reject the null hypothesis.

**Role in Deciding Test Outcome**

The significance level is the standard against which the test's result, the p-value, is compared to decide the outcome.

 **Comparing p-value and \(Î±):**

  After a statistical test, the p-value is calculated. The decision rule is as follows:

  - If the p-value \(Î±), the null hypothesis is rejected. The result is statistically significant, meaning there is strong evidence against the null hypothesis.

  - If the p-value (Î±), the null hypothesis is not rejected. The result is not statistically significant, meaning there is insufficient evidence to conclude that an effect exists.
  
**Influences test strictness:**
  
  The choice of (Î±) influences how strong the evidence must be to reject the null hypothesis.
  
  - **Lower (Î±) (e.g., 0.01):** A stricter test that reduces the risk of a Type I error. It requires stronger evidence (a smaller p-value) to reject the null hypothesis but increases the risk of a Type II error (a "false negative," where a false null hypothesis is not rejected).
  
  - **Higher (Î±) (e.g., 0.10):** A more lenient test that makes it easier to reject the null hypothesis. It increases the risk of a Type I error but decreases the risk of a Type II error.

**Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer-** In statistics, a Type I error, also known as a false positive, occurs when a true null hypothesis is incorrectly rejected. A Type II error, or false negative, happens when a false null hypothesis is not rejected

**Type I error (false positive)**

A Type I error is rejecting a true null hypothesis, concluding a significant finding when it's due to chance. The probability is denoted by **Î±**.

 - **Example: Clinical drug trial**

   - **(Hâ‚€)**: A new drug has no effect.
   - **(Hâ‚)** : The drug is effective.
   - **Type I Error**: Concluding the drug is effective when it is not. This could lead to approving an ineffective drug, potentially harming patients.

 - **Example: Airport security metal detector**

   - **(Hâ‚€)**: Item is not a weapon.
   - **(Hâ‚)**: Item is a weapon.
   - **Type I Error:** The detector alarms for a harmless item. This is a false alarm.

**Type II error (false negative)**

A Type II error is failing to reject a false null hypothesis, missing a real effect or difference. The probability is denoted by beta (\(\beta \)).

 - **Example: Clinical drug trial**

    - **(Hâ‚€)**: A new drug has no effect.
    - **(Hâ‚)**: The drug is effective.
    - **Type II Error:** Concluding the drug is not effective when it is. This means a beneficial drug wouldn't reach patients.

 - **Example: Criminal court trial**

   - **(Hâ‚€)**: Defendant is innocent.
   - **(Hâ‚):** Defendant is guilty.
   - **Type II Error:** Acquitting a defendant who is guilty, allowing a criminal to go free

**Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**


```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50   # hypothesized population mean
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Standard error
se = sample_std / np.sqrt(n)

# Z-statistic
z_stat = (sample_mean - mu_0) / se

# p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

**Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # true mean and standard deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Confidence interval using Z-distribution (large sample)
confidence = 0.95
z_value = stats.norm.ppf(1 - (1-confidence)/2)
ci_lower = sample_mean - z_value * se
ci_upper = sample_mean + z_value * se

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Step 3: Plot histogram of the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")

plt.title("Normal Distribution Simulation with 95% CI")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

**Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.**


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample std dev
    z_scores = [(x - mean) / std_dev for x in data]
    return np.array(z_scores)

# Example dataset
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7, density=True)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.show()
```
